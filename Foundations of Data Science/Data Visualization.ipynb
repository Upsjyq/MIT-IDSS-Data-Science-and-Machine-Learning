This is the notebook for the data visualization classes in the course

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_data = sns.load_dataset("tips")
tips_data.head()
```

## Line plot

```python
x = np.linspace(1, 20, 100)
y = np.exp(x)

plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, "r*")
plt.show()
```

```python
y1 = x
y2 = np.square(x)
y3 = np.sqrt(x)

plt.plot(x, y1, 'r')
plt.show()
```

```python
plt.plot(x, y2, 'b')
plt.show()
```

```python
plt.plot(x, y3, 'g')
plt.show()
```

```python
plt.plot(x, y1, 'r', x, y2, 'b', x, y3, 'g')
plt.title('line plot')
plt.xlabel('x')
plt.ylabel('y1, y2, y3')
plt.show()
```

## Scatter plot

```python
tips_data.head()
```

```python
x = tips_data['total_bill']
y = tips_data['tip']

plt.scatter(x, y)
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.title('tips X total bill', size=20)
plt.show()
```

```python
# s = shape, c = colour, alpha = opacity (0 < alpha < 1)
plt.scatter(x, y, s=np.array(y)**2, c='green', alpha=0.8)

plt.title('scatter plot', size=20)
plt.xlabel('total bill')
plt.ylabel('tip')

plt.show()
```

## Bar plot

```python
tips_data['smoker'].value_counts().plot(kind='bar')

plt.show()
```

adding text to the plot

```python
tips_data.smoker.value_counts().plot(kind='bar', rot=0, color='green')

# x and y are the positions
# s is the text
plt.text(x=-0.05, y=tips_data.smoker.value_counts()[1]+1, s=tips_data.smoker.value_counts()[1])
plt.text(x=0.98, y=tips_data.smoker.value_counts()[0]+2, s=tips_data.smoker.value_counts()[0])

plt.title('bar plot', size=20)
plt.xlabel('smoker')
plt.ylabel('amount')

plt.show()
```

## Pie plot

```python
tips_data.head()
```

```python
x = tips_data['day'].value_counts()
print(x)
```

```python
plt.pie(x)

plt.show()
```

adding things to the plot

```python
# autopct includes the percentage each part of the pie represents. To change the number of decimal points, change the number
# before de 'f'
plt.pie(x, autopct = '%.1f%%')

plt.show()
```

```python
# radius changes the value of the radius of the pie plot (bigger radius bigger pie)
plt.pie(x, autopct='%.1f%%', radius=1.2)

plt.show()
```

```python
# add labels to see what each part of the pie represents
plt.pie(x, autopct='%.1f%%', radius=1.2, labels=['sat','sun','thur','fri'])

plt.show()
```

```python
# detach a part of the pie
plt.pie(x, autopct='%.1f%%', radius=1.2, labels=['sat','sun','thur','fri'], explode=[0,0,0,0.5])

plt.show()
```

adding a circle in the middle

```python
plt.pie(x, autopct='%.1f%%', radius=1.2, labels=['sat','sun','thur','fri'])

circle = plt.Circle( (0,0), 0.5, color='white') # creates an object circle
plot = plt.gcf() # gets the current figure (pie) and links it to the object plot
plot.gca().add_artist(circle) # gets the current axes and adds the circle to the plot 

plt.show()
```

## Histogram

## Boxplots

```python
plt.boxplot(tips_data['total_bill'])

plt.show()
```

writing on the plot

```python
plt.boxplot(tips_data['total_bill'])

plt.text(x=1.1, y=tips_data['total_bill'].min(), s='min')
plt.text(x=1.1, y=tips_data['total_bill'].quantile(0.25), s='Q1')
plt.text(x=1.1, y=tips_data['total_bill'].median(), s='median (Q2)')
plt.text(x=1.1, y=tips_data['total_bill'].quantile(0.75), s='Q3')
plt.text(x=1.1, y=tips_data['total_bill'].max(), s='max')

plt.title('total bill boxplot', size=20)
plt.ylabel('total bill')

plt.show()
```

Indicating outliers

```python
plt.boxplot(tips_data['total_bill'])

plt.text(x=1.1, y=tips_data['total_bill'].min(), s='min')
plt.text(x=1.1, y=tips_data['total_bill'].quantile(0.25), s='Q1')
plt.text(x=1.1, y=tips_data['total_bill'].median(), s='median (Q2)')
plt.text(x=1.1, y=tips_data['total_bill'].quantile(0.75), s='Q3')
plt.text(x=1.1, y=tips_data['total_bill'].max(), s='max')

plt.annotate('Outliers', xy=(0.97, 45), xytext=(0.7, 44), arrowprops=dict(facecolor='black', arrowstyle='simple'))

plt.title('total bill boxplot', size=20)
plt.ylabel('total bill')

plt.show()
```

## Strip plot

```python
tips_data.head()
```

```python
sns.stripplot(y='tip', x='time', data=tips_data)
plt.show()
```

```python
# jitter parameter enables to see the spread of the data
sns.stripplot(y='tip', x='time', data=tips_data, jitter=True)

plt.show()
```

## Swarm Plot
exactly the same as the stripplot, but makes observation clearer

```python
sns.swarmplot(y='tip', x='time', data=tips_data)

plt.show()
```

Adding another categorical variable

```python
sns.swarmplot(y='tip', x='time', data=tips_data, hue='day')

plt.show()
```

## Violin plot
does the same as the swarmplot, but instead of points, shows the distribution of it

```python
sns.violinplot(y='total_bill', x='day', data=tips_data)

plt.show()
```

changing the size and adding another catagorical variable

```python
plt.figure(figsize=(10,8))

sns.violinplot(y='total_bill', x='day', data=tips_data, hue='sex', split=True)
# hue: third variable (must be categorical)
# split = True: splits the 'violins for the hue'

plt.show()
```

```python
plt.figure(figsize=(10,8))

sns.violinplot(y='total_bill', x='day', data=tips_data, hue='sex', split=False)
# hue: third variable (must be categorical)
# split = False: creates a violin for each category in hue

plt.show()
```

## Pair plot

```python
plt.figure(figsize=(10,8))

sns.pairplot(tips_data)

plt.show()
```

## Distribution plot

```python
sns.distplot(tips_data['total_bill'])

plt.show()
```

```python
sns.histplot(tips_data['total_bill'])

plt.show()
```

distplot over the time

```python
times = ['Lunch', 'Dinner']

for i in times:
    subset = tips_data[tips_data['time'] == i]
    sns.distplot(subset['total_bill'], hist=False, kde=True, kde_kws={'shade':True}, label=i)
```

## Count plot

```python
sns.countplot(data=tips_data, x='day', hue='time')

plt.show()
```
